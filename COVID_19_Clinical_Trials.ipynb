### **Dataset Overview**
- The dataset consists of 5783 entries and 27 columns. Below is a summary of the key columns and their descriptions:

**Column Analysis**
- **Rank, NCT Number, Title:** Metadata for each trial.

- **Phases:** Missing for 42% of rows; defines trial phases.
- **Enrollment:** Numeric, missing for 0.6% of rows; highly skewed.
- **Dates:** Includes Start Date, Primary Completion Date, and Completion Date.
- **Locations, Study Documents:** Location data is partially missing (10%), and Study Documents has significant missingness (96.8%).
- **URL:** Unique for every trial.

```python
def extract_age_range(age):
    if 'Years to' in age:
        age_range = age.split(' to ')
        return int(age_range[0].split(' ')[0]), int(age_range[1].split(' ')[0])

    elif 'and older' in age:
        min_age = int(age.split(' ')[0])
        return min_age, np.inf

    elif 'Child' in age:
        return 0, 18
    elif 'Adult' in age:
        return 18, 65
    elif 'Older Adult' in age:
        return 65, 100
    else:
        return np.nan, np.nan

data[['Age_min', 'Age_max']] = data['Age'].apply(lambda x: pd.Series(extract_age_range(x)))
print(data[['Age', 'Age_min', 'Age_max']].head(10))
```

```python
bins = [0, 18, 35, 60, 100]
labels = ['Child', 'Adolescent', 'Adult', 'Senior']

data['Age Category'] = pd.cut(data['Age_min'], bins=bins, labels=labels, right=False)
print(data['Age Category'].value_counts())

plt.figure(figsize=(6, 4))
sns.countplot(x=data['Age Category'], palette='bright')
plt.title('Distribution of Age Categories in Clinical Trials', fontsize=16)
plt.xlabel('Age Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.tight_layout()
plt.show()
```
